```python
from thinkdsp import CosSignal, SinSignal
from thinkdsp import decorate
import matplotlib.pyplot as plt

cos_sig = CosSignal(freq=10, amp=10, offset=0)
plt.grid(True)
cos_sig.plot()
decorate(xlabel='sec')
```

```python
sin_sig = SinSignal(freq=20, amp=10, offset=0)
plt.grid(True)
sin_sig.plot()
decorate(xlabel='sec')
```

```python
mix = sin_sig + cos_sig
plt.grid(True)
mix.plot()
decorate(xlabel='sec')
```

```python
wave = mix.make_wave(duration=2, start=0, framerate=11025)
plt.grid(True)
wave.plot()
decorate(xlabel='sec')
```

```python
from IPython.display import Audio
wave = mix.make_wave(duration=2, start=0, framerate=1000)
wave.make_audio()
print('samples',len(wave.ys))
```

```python
#1.6
spectrum = wave.make_spectrum()
plt.grid(True)
spectrum.plot(high=500)
decorate(xlabel='Hz')
```

```python
#1.7
sin1= SinSignal(freq=1000, amp=0.25)
cos1= CosSignal(freq=2000, amp=0.25)
sin2= SinSignal(freq=3000, amp=0.25)
cos2 = CosSignal(freq=4000, amp=0.25)
mix2=sin1+sin2+cos1+cos2
mix2.plot()
wave2=mix2.make_wave(duration=4)
```

```python
spec2=wave2.make_spectrum()
plt.grid()
spec2.plot()
decorate(xlabel='Hz')
```

```python
#1.8.1
spec2.high_pass(3000)
spec2.plot()
plt.grid()
decorate(xlabel='Hz')
```

```python
#1.8.2
spec2.low_pass(2000)
spec2.plot()
plt.grid()
decorate(xlabel='Hz')
```

```python
#2.2
from scipy import signal

import matplotlib.pyplot as plot

import numpy as np

 

t = np.linspace(1, 0, 1000, endpoint=False)

plot.plot(t, signal.square(-2 * np.pi * 10 * t))
plot.xlabel('sec')
plot.grid(True)
plot.show()
```

```python
#4.1
import numpy as np
import thinkdsp as td

a=10
sampleRate=1000
t=np.linspace(0, 2, 2*sampleRate+1)
x=np.cos(a * t**2)
w=td.Wave(x, framerate= sampleRate)
w.plot()
plt.grid()
```

```python
#4.3
spectrogram = w.make_spectrogram(seg_length=512)
spectrogram.plot(high=40)
```
